## Which country has the most people? 

```python
%%sql
SELECT  *,MAX(population)
FROM facts
WHERE name != 'World';
```

```python
%%sql
SELECT  *, MIN(population)
FROM facts
WHERE name != 'Antarctica';
```

```python
%%sql
SELECT *
FROM (SELECT name, population, death_rate-birth_rate AS  country_try_disappearing_ratio
FROM facts)
ORDER BY country_try_disappearing_ratio
```
